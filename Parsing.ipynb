# Парсим данные

Буду собирать информацию о двухкомнатных и трехкомнатных квартирах для долгосрочной аренды в Москве.

Парсим данные с ЦИАНа, использую уже готовую библиотеку https://github.com/lenarsaitov/cianparser

Чтобы не потерять все труды из-за возможных сбоев и ошибок, решила парсить по 50 страниц, поэтому буду собирать данные в три этапа. 

```python
import cianparser

moscow_parser = cianparser.CianParser(location="Москва")


def parse_with_delays(start_page, end_page, filename="cian_data.xlsx"):
    """
    Функция парсит объявления с Циан по заданному диапазону страниц, сохраняя данные в таблицу xlsx.

    Args:
        start_page (int): Номер первой страницы для парсинга.
        end_page (int): Номер последней страницы для парсинга.
        filename (str, optional): Имя файла xlsx для сохранения данных "cian_data.xlsx".
    """

    all_data = []  # Список для хранения данных со всех страниц

    for page in range(start_page, end_page + 1):
        try:
            data = moscow_parser.get_flats(
                deal_type="rent_long",
                rooms=(2, 3),
                additional_settings={
                    "start_page": page,
                    "end_page": page,  # Парсим по одной странице за раз
                    "metro_foot_minute": 10
                }
            )
            # Добавляем данные с текущей страницы в общий список
            all_data.extend(data)
            print(f"Страница {page} успешно обработана.")
        except Exception as e:
            print(f"Ошибка при обработке страницы {page}: {e}")

        # Случайная задержка между запросами
        delay = random.uniform(2, 5)  # Задержка от 2 до 5 секунд
        print(f"Задержка на {delay:.2f} секунд перед следующим запросом...")
        time.sleep(delay)

    # Создание таблицы из списка данных
    df = pd.DataFrame(all_data)

    # Проверка на наличие файла
    try:
        existing_data = pd.read_excel(filename)
    except FileNotFoundError:
        existing_data = []  # Если файл не существует, создаем пустой DataFrame

    # Объединение данных
    df_merged = pd.concat([existing_data, df], ignore_index=True)

    # Сохранение объединенных данных
    df_merged.to_excel(filename, index=False)

    # Вывод информации о таблице
    print(f"Собрано {df_merged.shape[0]} объявлений.")
    print(df_merged.columns)  # Показать названия столбцов

parse_with_delays(1, 50)  # Первая порция данных

# Все данные будут сохранены в файле cian_data.xlsx
```
